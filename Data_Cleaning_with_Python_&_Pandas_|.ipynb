```python
# Tekrar eden satırları kontrol et
df = df.drop_duplicates()
```

```python
df.loc[:,"name"]= df["name"].str.lower().str.title()
```

```python
age_mean = df["age"].mean().round(0)
```

```python
df.loc[:,"age"]=df["age"].fillna(age_mean)
```

```python
df.loc[:,"gender"] = df["gender"].str.lower().str.title()
```

```python
df["grade"].unique()
```

```python
df.loc[:,"grade"] = df["grade"].replace({'10th': '10', '11th': '11'})
```

```python
math_mean = df["math_score"].mean()
```

```python

df.loc[:, "math_score"] = df["math_score"].fillna(math_mean)


```

```python

import pandas as pd
import numpy as np

# 30 sətirlik DataFrame
n = 30
df = pd.DataFrame({
    "student_id": range(1, n+1),
    "name": [f"Student_{i}" for i in range(1, n+1)],
    "age": np.random.randint(15, 20, size=n),
    "gender": np.random.choice(["Male", "Female"], size=n),  # tam sözlə
    "grade": np.random.choice(["9th", "10th", "11th"], size=n),
    "math_score": np.random.randint(50, 100, size=n),
    "english_score": np.random.randint(60, 100, size=n),
    "science_score": np.random.randint(60, 100, size=n),
    "enrolled_date": [f"2025:{str(np.random.randint(1,13)).zfill(2)}:{str(np.random.randint(1,29)).zfill(2)}" for _ in range(n)],
    "remarks": np.random.choice(["good", "improvement", "needs review"], size=n)
})

print(df)
```
